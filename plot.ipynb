```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fileCsv = pd.read_csv(fileDir+fileName)
```

```python
def plotXY(csvDict,*args):
    if(type(csvDict)!= pd.core.frame.DataFrame):
        raise ValueError("csvFile is not a DataFrame\n")
    m_columns = []
    for x in args:
        if x in csvDict.columns:
            m_columns.append(x);
    if(len(m_columns)>0):
        out = pd.DataFrame(csvDict,columns = m_columns)
        out.plot()
    else:
        raise ValueError("input a column to plot\n")
```

```python
plotXY(fileCsv,"ReceiveRate")
```
